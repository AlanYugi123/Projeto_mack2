```python
import pandas as pd
import numpy as np

# Definindo o número de alunos por turma e série
num_alunos = 30

# Criando uma lista para armazenar os dados
dados = []
id = 0

# Gerando nomes brasileiros aleatórios (lista simplificada)
nomes = [
    'Ana', 'Lucas', 'Miguel', 'Arthur', 'Gabriel', 'Julia', 'Sophia', 'Maria', 'Alice', 'Heitor',
    'Pedro', 'Laura', 'Isabella', 'Manuela', 'Luiza', 'Lorena', 'Giovanna', 'Beatriz', 'Mariana', 'Gabriela'
]

# Gerando dados para cada turma e cada ano
for turma in ['A', 'B']:
    for ano in range(1, 6):
        for _ in range(num_alunos):
            nome = np.random.choice(nomes)
            matricula = np.random.randint(10000, 99999)  # Número de matrícula aleatório
            dados.append({
                'id': id,
                'nome': nome,
                'matricula': matricula,
                'turma': f'Turma {turma}',
                'ano': f'{ano}º Ano'
            })
            id += 1

# Criando o DataFrame
df = pd.DataFrame(dados)

df['id'] = df['id'].astype('int64')  # Garante que id é int64, mas geralmente já é
df['nome'] = df['nome'].astype('object')  # Garante que nome é tratado como objeto
df['matricula'] = df['matricula'].astype('object')  # Converte matricula para object
df['turma'] = df['turma'].astype('object')  # Converte turma para object
df['ano'] = df['ano'].astype('object')  # Converte ano para object

# Exibindo o DataFrame completo
print(df)
```

```python
# Lista de sobrenomes brasileiros comuns
sobrenomes = [
    'Silva', 'Santos', 'Oliveira', 'Souza', 'Rodrigues', 'Fernandes', 'Almeida', 'Costa', 'Lima', 'Gomes',
    'Pereira', 'Rocha', 'Martins', 'Araújo', 'Ribeiro', 'Barros', 'Alves', 'Cardoso', 'Schmidt', 'Carmo'
]

# Atualizando a coluna 'nome' para incluir um sobrenome aleatório
df['nome'] = df['nome'].apply(lambda x: x + ' ' + np.random.choice(sobrenomes))
```
